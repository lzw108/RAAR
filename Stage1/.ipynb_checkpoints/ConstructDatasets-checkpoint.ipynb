# AMTCele

```python
import pandas as pd

data_type = "train"

df_Vreg = pd.read_json("./retrieval/AMTCele/Vreg/%s/AMTCele_%s.json"%(data_type,data_type), lines=True)
df_Semantic = pd.read_json("./retrieval/AMTCele/roberta/%s/AMTCele_%s.json"%(data_type,data_type), lines=True)
df_Style = pd.read_json("./retrieval/AMTCele/styleembedding/%s/AMTCele_%s.json"%(data_type,data_type), lines=True)
```

```python
df_Vreg
```

```python
import re
import json

with open("./retrieval/AMTCele/data4filter/AMTCele_train4filter.json", 'w', encoding="utf-8") as write_f:


    for (index, vreg_row), (_, semantic_row), (_, style_row) in zip(df_Vreg.iterrows(),
                                                df_Semantic.iterrows(),
                                                df_Style.iterrows()):
        # traverse vreg_row.colname
    #     print(vreg_row, semantic_row, style_row)
        data_one = {}
        data_one["index"] = "AMTCele_train_%s"%index
        data_one["introduction"] = "Based on the following information to make judgements, just reply with the option [fake or legit]. DO NOT EXPLAIN:\n"
        data_one["task_description"] = "Task: Determine whether the target text is fake or legit.\n"
        data_one["target"] = vreg_row["Target"]
        data_one["fewshot_vreg"] = vreg_row["Prompt4fewshots"] + " ".join(vreg_row["Fewshots"][:2])
        data_one["fewshot_semantic"] = semantic_row["Prompt4fewshots"] + " ".join(semantic_row["Fewshots"][:2])
        data_one["fewshot_style"] = style_row["Prompt4fewshots"] + " ".join(style_row["Fewshots"][:2])
    #     data_one["end_instruction"] = 
        data_one["complete_instruction"] = data_one["introduction"] + data_one["task_description"] + data_one["target"] + data_one["fewshot_vreg"] + "\n" + data_one["fewshot_semantic"] + "\n" + data_one["fewshot_style"]     
        data_one["label"] = re.sub(r"^\d+\.\s*", "", vreg_row["Answer"])
        if vreg_row["Answer"] == "fake":
            data_one["response"] = "The target content is fake"
        elif vreg_row["Answer"] == "legit":
            data_one["response"] = "The target content is legit"

        write_f.write(json.dumps(data_one, indent=None, ensure_ascii=False) + "\n")
```

# PHEME

```python
import pandas as pd

data_type = "train"
data_name = "PHEME"

df_Vreg = pd.read_json("./retrieval/%s/Vreg/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
df_Semantic = pd.read_json("./retrieval/%s/roberta/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
df_Style = pd.read_json("./retrieval/%s/styleembedding/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
```

```python
df_Vreg
```

```python
import re
import json

with open("./retrieval/%s/data4filter/%s_train4filter.json"%(data_name,data_name), 'w', encoding="utf-8") as write_f:


    for (index, vreg_row), (_, semantic_row), (_, style_row) in zip(df_Vreg.iterrows(),
                                                df_Semantic.iterrows(),
                                                df_Style.iterrows()):

        data_one = {}
        data_one["index"] = "%s_%s_%s"%(data_name, data_type, index)
        data_one["introduction"] = "Based on the following information to make judgements, just reply with the option [false rumour, true rumour, unverified rumour, non-rumour]. DO NOT EXPLAIN:\n"
        data_one["task_description"] = "Task: Determine which one of the following four categories the target text belongs to: false rumour, true rumour, unverified rumour, or non-rumour.\n"
        data_one["target"] = vreg_row["Target"]
        data_one["fewshot_vreg"] = vreg_row["Prompt4fewshots"] + " ".join(vreg_row["Fewshots"][:2])
        data_one["fewshot_semantic"] = semantic_row["Prompt4fewshots"] + " ".join(semantic_row["Fewshots"][:2])
        data_one["fewshot_style"] = style_row["Prompt4fewshots"] + " ".join(style_row["Fewshots"][:2])
    #     data_one["end_instruction"] = 
#         data_one["complete_instruction"] = data_one["introduction"] + data_one["task_description"] + data_one["target"] + data_one["fewshot_vreg"] + "\n" + data_one["fewshot_semantic"] + "\n" + data_one["fewshot_style"]     
        data_one["complete_instruction"] = data_one["task_description"] + data_one["target"] + data_one["fewshot_vreg"] + "\n" + data_one["fewshot_semantic"] + "\n" + data_one["fewshot_style"]     
        data_one["label"] = re.sub(r"^\d+\.\s*", "", vreg_row["Answer"])
        if vreg_row["Answer"] == "false rumour":
            data_one["response"] = "The target content is false rumour"
        elif vreg_row["Answer"] == "true rumour":
            data_one["response"] = "The target content is true rumour"
        elif vreg_row["Answer"] == "unverified rumour":
            data_one["response"] = "The target content is unverified rumour"
        elif vreg_row["Answer"] == "non-rumour":
            data_one["response"] = "The target content is non-rumour"

        write_f.write(json.dumps(data_one, indent=None, ensure_ascii=False) + "\n")
```

# COCO

```python
import pandas as pd

data_type = "train"
data_name = "COCO"

df_Vreg = pd.read_json("./retrieval/%s/Vreg/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
df_Semantic = pd.read_json("./retrieval/%s/roberta/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
df_Style = pd.read_json("./retrieval/%s/styleembedding/%s/%s_%s.json"%(data_name,data_type,data_name,data_type), lines=True)
```

```python
import re
import json

with open("./retrieval/%s/data4filter/%s_train4filter.json"%(data_name,data_name), 'w', encoding="utf-8") as write_f:


    for (index, vreg_row), (_, semantic_row), (_, style_row) in zip(df_Vreg.iterrows(),
                                                df_Semantic.iterrows(),
                                                df_Style.iterrows()):

        data_one = {}
        data_one["index"] = "%s_%s_%s"%(data_name, data_type, index)
        data_one["introduction"] = "Based on the following information to make judgements, just reply with the option [Unrelated, Related, Conspiracy]. DO NOT EXPLAIN:\n"
        data_one["task_description"] = "Task: Determine whether the target text is Unrelated, Related, or Conspiracy. Unrelated means the text contains conspiracy-related keywords but uses them in an unrelated or different context. Related means the text is conspiracy-related but does not propagate conspiracy misinformation. Conspiracy means the text is conspiracy-related and actively propagates or supports the misinformation.\n"
        data_one["target"] = vreg_row["Target"]
        data_one["fewshot_vreg"] = vreg_row["Prompt4fewshots"] + " ".join(vreg_row["Fewshots"][:2])
        data_one["fewshot_semantic"] = semantic_row["Prompt4fewshots"] + " ".join(semantic_row["Fewshots"][:2])
        data_one["fewshot_style"] = style_row["Prompt4fewshots"] + " ".join(style_row["Fewshots"][:2])
    #     data_one["end_instruction"] = 
        data_one["complete_instruction"] = data_one["introduction"] + data_one["task_description"] + data_one["target"] + data_one["fewshot_vreg"] + "\n" + data_one["fewshot_semantic"] + "\n" + data_one["fewshot_style"]     
        data_one["label"] = re.sub(r"^\d+\.\s*", "", vreg_row["Answer"]).split("(")[0].strip()
        if vreg_row["Answer"] == "Related (but not supporting)":
            data_one["response"] = "The target content is Related, which is conspiracy-related but does not propagate misinformation"
        elif vreg_row["Answer"] == "Conspiracy (related and supporting)":
            data_one["response"] = "The target content is Conspiracy, which is conspiracy-related and propagates conspiracy misinformation"
        elif vreg_row["Answer"] == "Unrelated":
            data_one["response"] = "The target content is Unrelated, which contains conspiracy-related keywords but uses them in an unrelated or different context."


        write_f.write(json.dumps(data_one, indent=None, ensure_ascii=False) + "\n")
```
