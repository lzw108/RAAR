# semantic

### semantic

```python
import numpy as np
import pandas as pd
import os
import torch

from text2vec import SentenceModel


dataname = "AMTCele"
save_type = "roberta"

# Roberta Senmantic
model = SentenceModel("roberta-base")

if dataname == "AMTCele":
    CSV_FILE = os.path.join("../datasets/AMTCele/AMTcele.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['content'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/AMTCele/AMTCele_%s_embeddings.pth"%save_type)
elif dataname == "PHEME":
    CSV_FILE = os.path.join("../datasets/PHEME/PHEME.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['text'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/PHEME/PHEME_%s_embeddings.pth"%save_type)
elif dataname == "COCO":
    CSV_FILE = os.path.join("../datasets/COCO/COCO.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['Tweet'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/COCO/COCO_%s_embeddings.pth"%save_type)
    
print("Finished")
```

# Style

### style-embedding (https://huggingface.co/AnnaWegmann/Style-Embedding)

```python
import numpy as np
import pandas as pd
import os
import torch

from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

dataname = "COCO"
save_type = "styleembedding"

model = SentenceTransformer("AnnaWegmann/Style-Embedding")

if dataname == "AMTCele":
    CSV_FILE = os.path.join("../datasets/AMTCele/AMTcele.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['content'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/AMTCele/AMTCele_%s_embeddings.pth"%save_type)
elif dataname == "PHEME":
    CSV_FILE = os.path.join("../datasets/PHEME/PHEME.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['text'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/PHEME/PHEME_%s_embeddings.pth"%save_type)
elif dataname == "COCO":
    CSV_FILE = os.path.join("../datasets/COCO/COCO.csv")
    metadata = pd.read_csv(CSV_FILE)
    corpus_name_summary = metadata['Tweet'].tolist()
    print("In encoding......")
    corpus_embeddings = model.encode(corpus_name_summary)
    torch.save(corpus_embeddings, "../datasets/COCO/COCO_%s_embeddings.pth"%save_type)
    
print("Finished")
```
